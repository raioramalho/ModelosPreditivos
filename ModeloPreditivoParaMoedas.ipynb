```python
import gradio as gr
import pandas as pd
import requests
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
# URL da API
url = "https://economia.awesomeapi.com.br/last/USD-BRL,EUR-BRL,BTC-BRL"

# Faz a requisição à API
response = requests.get(url)

df = []

# Verifica se a requisição foi bem-sucedida (código de status 200)
if response.status_code == 200:
    # Converte os dados JSON em um dicionário
    data = response.json()

    # Cria listas para armazenar os dados
    currencies = []
    highs = []
    lows = []
    bids = []
    asks = []

    # Itera sobre as chaves do dicionário
    for key in data.keys():
        currencies.append(data[key]['name'])
        highs.append(float(data[key]['high']))
        lows.append(float(data[key]['low']))
        bids.append(float(data[key]['bid']))
        asks.append(float(data[key]['ask']))

    # Cria um DataFrame
    df = pd.DataFrame({
        'Moeda': currencies,
        'Alta': highs,
        'Baixa': lows,
        'Bid': bids,
        'Ask': asks
    })

    # Exibe o DataFrame
    print("Requisição feita com sucesso!")

else:
    print(f"Erro na requisição. Código de status: {response.status_code}")
```
