```python
import gradio as gr
import pandas as pd
```

```python
# prompt: importe as libs para trabalhar com modelo de arvore de decisao sklearn

from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
# prompt: gere 3 arrays com 20 itens cada, idade peso e altura com valores humanizados, valores manuais, sem usar random

idade = [18, 22, 26, 30, 34, 38, 42, 46, 50, 54, 58, 62, 66, 70, 74, 78, 82, 86, 90, 94]
peso = [60, 65, 70, 75, 180, 85, 90, 195, 100, 105, 210, 115, 120, 125, 130, 135, 140, 145, 250, 155]
altura = [1.50, 1.55, 1.60, 1.65, 1.70, 1.75, 1.80, 1.85, 1.90, 1.95, 2.00, 2.05, 2.10, 2.15, 2.20, 2.25, 2.30, 2.35, 2.40, 2.45]
```

```python
# prompt: gera um array chamado rotulo para possiveis status de obesidade de acordo com os valores de idade peso e altura

rotulo = []
for i in range(len(idade)):
  if peso[i] / (altura[i]**2) > 30:
    rotulo.append("Obeso")
  else:
    rotulo.append("Nao obeso")
```

```python
rotulo
```

```python
# prompt: crie um dataframe com as colunas e os respectivos valores de idade, altura, peso e rotulo

df = pd.DataFrame({'idade': idade, 'altura': altura, 'peso': peso, 'rotulo': rotulo})
```

```python
df
```

```python
# prompt: crie um modelo baseado em arvore de decisao para receber uma idade altura e peso que retorne o possivel rotulo

def predict_obesity(idade, altura, peso):
  gordo = "https://png.pngtree.com/thumb_back/fw800/background/20230605/pngtree-the-fat-man-s-pose-has-the-big-belly-image_2884415.jpg"
  magro = "https://melhorcomsaude.com.br/wp-content/uploads/2021/11/musculo-delgado-500x333-1.jpg"
  # Create a decision tree classifier
  clf = DecisionTreeClassifier()

  # Split the data into training and test sets
  X_train, X_test, y_train, y_test = train_test_split(df[['idade', 'altura', 'peso']], df['rotulo'], test_size=0.25)

  # Train the classifier
  clf.fit(X_train, y_train)

  # Predict the obesity status of the given person
  y_pred = clf.predict([[idade, altura, peso]])[0]

  # Return the prediction
  # if y_pred == 'Obeso':
  #   return gordo
  # else:
  #   return magro
  return y_pred

# Create a Gradio interface for the model
# interface = gr.Interface(fn=predict_obesity, inputs=[gr.Slider(label="Idade"), gr.Number(label="Altura"), gr.Slider(label="Peso", maximum=300)], outputs=gr.Image(label="Prediction"))

# Launch the interface
# interface.launch()

predict_obesity(26, 1.67, 80)
```

```python
# prompt: exponha uma rota http que receba um post com idade peso e altura

from flask import Flask, request, jsonify

app = Flask(__name__)

@app.route('/predict', methods=['POST'])
def predict():
  # Get the request data
  data = request.get_json()

  # Get the age, height, and weight from the data
  age = data['idade']
  height = data['alutra']
  weight = data['peso']

  # Predict the obesity status
  prediction = predict_obesity(age, height, weight)

  # Return the prediction
  return jsonify({'prediction': prediction})

if __name__ == '__main__':
  app.run()
```
